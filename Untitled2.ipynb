```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("data/chinook.db")
#sales
#by month
sqlStat_salesByMonth = "SELECT * FROM INVOICES order by invoiceDate"
albums = pd.read_sql_query(sqlStat_salesByMonth, conn,parse_dates='InvoiceDate')
albums['bulan'] = albums['InvoiceDate'].dt.month_name()
albums
salesbyMonth = albums.groupby('bulan').sum()
salesbyMonthdf = pd.DataFrame(salesbyMonth,columns=['Total'])
salesbyMonthdf.stack()
# salesbyMonthdf.xs('Total',level='bulan',axis=0)
# js =  salesbyMonthdf.to_json()
# js
```
